```python
population_df = world_pop.copy()
```

```python
#converting YearlyChange, Urbanpop% and WorldShare to numeric columns

columns_to_convert = ['YearlyChange', 'UrbanPop%', 'WorldShare']
```

```python
for column in columns_to_convert:
    population_df[column] = pd.to_numeric(world_pop[column].str.replace('%', ''), errors='coerce')
```

```python
 population_df.isna().sum() / len(population_df) * 100
```

```python
population_df['UrbanPop%'].fillna(population_df['UrbanPop%'].mean(), inplace=True)
```

```python
population_df.columns
```

```python
### Check for outliers

outlier_df = population_df[['Rank', 'Population2023', 'YearlyChange', 'NetChange',
       'Density(P/Km²)', 'Land Area(Km²)', 'Migrants(net)', 'Fert.Rate',
       'MedianAge', 'UrbanPop%', 'WorldShare']]
```

```python
from scipy.stats import zscore

# Calculate Z-scores for Population2023
z_scores = zscore(outlier_df['Population2023'])

# Identify outliers (using a threshold, e.g., Z-score > 3 or < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Display rows with outliers
outliers_df = outlier_df[outliers]
print("Rows with Outliers:")
print(outliers_df)
```

```python
# Calculate Z-scores for Population2023
z_scores = zscore(outlier_df['Density(P/Km²)'])

# Identify outliers (using a threshold, e.g., Z-score > 3 or < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Display rows with outliers
density_df = outlier_df[outliers]
print("Rows with Outliers:")
print(density_df)
```

```python
#remove outliers

density_df_no_outliers = outlier_df[~outliers]

# Display the DataFrame without outliers
print("DataFrame without Outliers:")
print(density_df_no_outliers)
```

```python
# Create subplots with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Scatter plots for UrbanPop% against other demographic factors
sns.lineplot(x='UrbanPop%', y='Population2023', data=population_df, ax=axes[0, 0],color='red',)
axes[0, 0].set_title('UrbanPop% vs. Population2023')

sns.lineplot(x='UrbanPop%', y='Density(P/Km²)', data=population_df, ax=axes[0, 1],color='blue',)
axes[0, 1].set_title('UrbanPop% vs. Density(P/Km²)')

sns.lineplot(x='UrbanPop%', y='Fert.Rate', data=population_df, ax=axes[1, 0],color='green',)
axes[1, 0].set_title('UrbanPop% vs. Fert.Rate')

sns.lineplot(x='UrbanPop%', y='MedianAge', data=population_df, ax=axes[1, 1],color='purple',)
axes[1, 1].set_title('UrbanPop% vs. MedianAge')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python

top5_migrants = population_df.sort_values(by='Migrants(net)', ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top5_migrants['Country'], top5_migrants['Migrants(net)'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Number of Migrants')
plt.title('Top 5 Countries with Most Migrants')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```
